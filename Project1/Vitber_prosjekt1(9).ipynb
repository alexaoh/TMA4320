# Exercise 3.1: 
We will show that equation (6) in the task description is a solution of the diffusion equation by simply inserting it into the equation with constant diffusion coefficient, since the coefficient $D$ is independent of x in the solution. 

To show that 

$\phi(x,t) = \frac{1}{\sqrt{4\pi Dt}} \int^{+\infty}_{-\infty}dy\,h(y)e^{-\frac{(x-y)^2}{4Dt}}$

is a solution of

$\frac{\partial \phi(x, t)}{\partial t} = D\frac{\partial^2\phi(x,t)}{\partial^2x}$,

we first calculate

$\frac{\partial \phi(x, t)}{\partial t} = -\frac{1}{2} \frac{1}{\sqrt{4\pi Dt}} \frac{1}{t} \int^{+\infty}_{-\infty}dy\,h(y)e^{-\frac{(x-y)^2}{4Dt}} + \frac{1}{\sqrt{4\pi Dt}} \frac{1}{t^2}\int^{+\infty}_{-\infty}dy\,h(y) \, \frac{(x-y)^2}{4D}e^{-\frac{(x-y)^2}{4Dt}} = \frac{1}{\sqrt{4\pi Dt}}(\frac{1}{t^{2}} \int^{+\infty}_{-\infty}dy\,h(y)\,\frac{(x-y)^2}{4D}e^{-\frac{(x-y)^2}{4Dt}} - \frac{1}{2t} \int^{+\infty}_{-\infty}dy\,h(y)e^{-\frac{(x-y)^2}{4Dt}})$.

Then we find

$\frac{\partial\phi(x,t)}{\partial x} = \frac{1}{\sqrt{4\pi Dt}} (-\frac{1}{4Dt}) \int^{+\infty}_{-\infty}dy\,h(y)\,(2x-2y)e^{-\frac{(x-y)^2}{4Dt}}$

$\implies \frac{\partial^2 \phi(x,t)}{\partial^2 x} = \frac{1}{\sqrt{4\pi Dt}}\frac{1}{4Dt}(-\int^{+\infty}_{-\infty}dy\,h(y)\,2e^{-\frac{(x-y)^2}{4Dt}} + \int^{+\infty}_{-\infty}dy\,h(y)\,\frac{(2x-2y)^2}{4Dt}e^{-\frac{(x-y)^2}{4Dt}}) = \frac{1}{D\sqrt{4\pi Dt}}(\frac{1}{t^{2}} \int^{+\infty}_{-\infty}dy\,h(y)\,\frac{(x-y)^2}{4D}e^{-\frac{(x-y)^2}{4Dt}} - \frac{1}{2t} \int^{+\infty}_{-\infty}dy\,h(y)e^{-\frac{(x-y)^2}{4Dt}})$.

It is apparent that the expression (6) is a solution of the equation:

$\implies \frac{\partial \phi(x, t)}{\partial t} = D\frac{\partial^2\phi(x,t)}{\partial^2x}$ $$\tag*{$\blacksquare$}$$


By letting $t \longrightarrow 0$ in Eq. (6) it becomes apparent that the distribution $\phi (x,t)$ becomes narrower. It reduces to a very sharp line (quite like Dirac's Delta) around x = y. Therefore, one may conclude that $h(y)$ represents the initial distribution of the particles. 

This solution (6) of the diffusion equation can also be found by applying a Fourier transform to the diffusion equation. By finding the solution this way, it becomes apparent from the beginning of the calculations that the function $h$ represents the initial distribution of the particles. # Exercise 3.2:
The probability of finding a particle in position $x$ in a potential landscape $V(x)$ is given by $P(x) = Ze^{-\beta V(x)}$, where $Z$ is the canonical partition function of the system and $\beta = (k_B T)^{-1}$.

During a random walk the probability of a particle moving in the positive direction, staying put or moving in the negative direction is, respectively, $\{p^{+}, p^{0}, p^{-}\}$. Normalization dictates that their combined sum should equal one. This means the probabilities must satisfy the equation 

\begin{equation}\label{eq1}\tag{1}
p^{+} + p^{0} + p^{-} = 1.
\end{equation}

We assume that each of the probabilities are proportional to $P(x)$ and that the proportionality constant, $C$, is equal for all steps. If the position before a random walk step is $x_0$ we get 

\begin{equation}
  \begin{aligned}
    p^{+} & =CP(x_0+1),\\
    p^{0} & =CP(x_0),\\
    p^{-} & =CP(x_0-1). 
    \label{eq2}
  \end{aligned}
\end{equation}

By putting these expressions into equation \eqref{eq1} we find the constant 

\begin{equation}\label{eq3}
C=\frac{1}{P(x_0+1)+P(x_0)+P(x_0-1)},
\end{equation}

and the desired probabilities

\begin{equation}
  \begin{aligned}
    p^{+} &=\frac{P(x_0+1)}{P(x_0+1)+P(x_0)+P(x_0-1)} = \frac{1}{1+e^{-\beta [V(x_0-1)-V(x_0+1)]}+e^{-\beta [V(x_0)-V(x_0+1)]}}, \\
    p^{0} &=\frac{P(x_0)}{P(x_0+1)+P(x_0)+P(x_0-1)} = \frac{1}{1+e^{-\beta [V(x_0-1)-V(x_0)]}+e^{-\beta [V(x_0+1)-V(x_0)]}}, \\
    p^{-} &=\frac{P(x_0-1)}{P(x_0+1)+P(x_0)+P(x_0-1)} = \frac{1}{1+e^{-\beta [V(x_0+1)-V(x_0-1)]}+e^{-\beta [V(x_0)-V(x_0-1)]}}.
    \label{eq4}
  \end{aligned}
\end{equation}
$$\tag*{$\blacksquare$}$$

# Exercise 3.3:
With the condition

$k_B T  \gg  \left|V(x+1)-V(x) \right| = \left|\Delta V(x)\right|  \implies  \beta \left|\Delta V(x) \right| \ll 1,$

all the exponential terms will reduce to one in the first order series expansion of $\exp(x)$;  $e^{-\beta \left|\Delta V(x) \right|} \approx e^{\beta \left| \Delta V(x) \right|} \approx 1$. Since the condition must hold for any $x$, the difference in potential, $\left| \Delta V(x) \right|$, represents a difference for an arbitrary position $x$. By inserting $\left| \Delta V(x) \right|$ in the probabilities found in task 3.2 and taking the absolute value we get

\begin{equation}\label{eq5}\tag{2}
p^{+} = \frac{1}{1+e^{2 \beta \left| \Delta V(x) \right|} + e^{\beta \left| \Delta V(x) \right|}} \approx \frac{1}{3},
\end{equation}

\begin{equation}\label{eq6}\tag{3}
p^{0} = \frac{1}{1+e^{\beta \left| \Delta V(x) \right|} + e^{-\beta \left| \Delta V(x) \right|}} \approx \frac{1}{3}, 
\end{equation}

\begin{equation}\label{eq7}\tag{4}
p^{-} = \frac{1}{1+e^{-2 \beta \left| \Delta V(x) \right|} + e^{-\beta \left| \Delta V(x) \right|}} \approx \frac{1}{3}.
\end{equation}

This seams rational since the potential is insignificant in relation to the thermal energy with the mentioned condition and therefore the particles must have equal probabilities to either stay put, move to left or to the right. 

The other condition states 
$k_B T  \ll  \left|V(x+1)-V(x) \right| = \left|\Delta V(x)\right|$. From this we obatin 
$\beta \left|\Delta V(x) \right| \gg 1 \implies e^{\beta \left| \Delta V(x) \right|} \approx \infty,$ and for the negative case 
$ -\beta \left|\Delta V(x) \right| \ll 1 \implies e^{-\beta \left| \Delta V(x) \right|} \approx 0$. 

With the same expressions for the probabilities as in equation \eqref{eq5}, \eqref{eq6} and \eqref{eq7} it becomes evident that $p^{+} \approx p^{0} \approx 0$ and

\begin{equation}
p^{-} = \frac{1}{1+e^{-2 \beta \left| \Delta V(x) \right|} + e^{-\beta \left| \Delta V(x) \right|}} \approx 1.
\end{equation}

If the terms in the condition were arranged the opposite way, $k_B T  \ll  \left|V(x)-V(x+1) \right|$, the probability $p^{+}$ would instead go towards one and $p^{0} \approx p^{-} \approx 0$. 

This shows that in the first case the potential is negligible, and in the second the thermal energy is neglible. 

These approximations show that if the ratio between the potential and the thermal energy is high the thermal energy makes a negligible impact on the result compared to the potential energy. The energy scale of the potential energy is much larger than that of the thermal energy. Similarly, the potential energy makes a negligible impact on the results when the ratio is small. We shall deal with this ratio, $\beta k$, later.

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt 
from math import e
import random as rand
from numba import jit

@jit(nopython=True)
def p_plus(x_0, V, beta_k):
    '''
    Probability of random walk step to the right.
    '''
    return 1/(1+e**(-beta_k*(V(x_0-1)-V(x_0+1)))+e**(-beta_k*(V(x_0)-V(x_0+1))))

@jit(nopython=True)
def p_zero(x_0, V, beta_k):
    '''
    Probability of staying put during random walk simulation.
    '''
    return 1/(1+e**(-beta_k*(V(x_0-1)-V(x_0)))+e**(-beta_k*(V(x_0+1)-V(x_0))))

@jit(nopython=True)
def p_minus(x_0, V, beta_k):
    '''
    Probability of random walk step to the left.
    '''
    return 1/(1+e**(-beta_k*(V(x_0+1)-V(x_0-1)))+e**(-beta_k*(V(x_0)-V(x_0-1))))

#Constants used in exercise 4
time_steps = 200
amount_of_particles = 10000 
beta_k = [0.01, 1, 100] 
step_length = 1
initial = np.zeros((amount_of_particles,), dtype=np.int64)

#Brukes denne n√•?!
@jit(nopython=True)
def get_result(V, betaAndK, init):
    '''
    Returns an array of the particles after being influenced by the potential V and the ratio BetaK. 
    Randomly decided if the particles move left, right or stay put. 
    '''
    new_array = init.copy()
    for step in range(time_steps):
        for position in range(len(init)):
            p = np.random.uniform(0,1)
            if p <= p_minus(new_array[position],V, betaAndK):
                new_array[position] -= step_length
            elif p > (1-p_plus(new_array[position],V, betaAndK)):
                new_array[position] += step_length
    return new_array

@jit(nopython=True) 
def mean(data):
    '''
    Finds the mean of data. 
    '''
    sum = 0
    for element in data:
        sum += element
    return sum/len(data)

@jit(nopython=True)
def std(data):
    '''
    Finds the standard deviation of data. 
    '''
    sum = 0
    mu = mean(data)
    n = len(data)
    for element in data:
        sum += (element-mu)**2
    return np.sqrt(sum/(n-1))

@jit(nopython=True)
def normal(x,mean,std):
    '''
    Finds the normal distribution. 
    '''
    return 1/(std*np.sqrt(2*np.pi)) * np.e**(-1/2 * ((x-mean)/std)**2)


def plot_hist(result1, result2, result3):  
    '''
    Histogram plot.
    '''
    x1 = np.arange(min(result1),max(result1),1)
    x2 = np.arange(min(result2),max(result2),1)
    x3 = np.arange(min(result3),max(result3),1)
    x_values = [x1, x2, x3]
    xaxe = "Particle position"
    yaxe = "Intensity"
    titles = [r'$\beta k = 0.01$',r'$\beta k = 1$',r'$\beta k = 100$'] 
    data = [result1, result2, result3]
    f,a = plt.subplots(1,3, figsize=(15,6))
    a[0].set_ylabel(yaxe, fontsize=20)
    a = a.ravel()
    for idx,ax in enumerate(a):
        ax.hist(data[idx])
        ax.set_title(titles[idx], fontsize=20)
        ax.set_xlabel(xaxe, fontsize=20)
        ax.text(0.5,-0.2, "Mean: " + str(mean(data[idx])) + ", Std: " + str(round(std(data[idx]),3)) , size=15, ha="center",transform=ax.transAxes)
    plt.tight_layout(pad = 2.0)
    plt.show()

def plot_normhist(result1, result2, result3): 
    '''
    Plots normalized histogram plot. 
    '''
    x1 = np.arange(min(result1),max(result1),1)
    x2 = np.arange(min(result2),max(result2),1)
    x3 = np.arange(min(result3),max(result3),1)
    x_values = [x1, x2, x3]
    xaxe = "Particle position"
    yaxe = "PDF"
    titles = [r'$\beta k = 0.01$',r'$\beta k = 1$',r'$\beta k = 100$'] 
    data = [result1, result2, result3]
    f,a = plt.subplots(1,3, figsize=(15,6))
    #a = a.ravel()
    a[0].set_ylabel(yaxe, fontsize=20)
    for idx,ax in enumerate(a):
        ax.hist(data[idx], density = True, label="PDF")
        ax.set_title(titles[idx], fontsize=20)
        ax.set_xlabel(xaxe, fontsize=20)
        ax.text(0.5,-0.2, "Mean: " + str(mean(data[idx])) + ", Std: " + str(round(std(data[idx]),3)) , size=15, ha="center", 
         transform=ax.transAxes)
        if (std(data[idx]) != 0):
            ax.plot(x_values[idx], normal(x_values[idx], mean(data[idx]), std(data[idx])))
        
    plt.tight_layout(pad = 2.0)
    plt.show()
```

```python
#Constants
h = 1   #step length
T_p = 500 
delta_t = 1
beta_k = 1000
N_x = 90  
N_p = 12*N_x
cycles = 10
jumps = 2*cycles

alpha = 0.8

x_first=-N_x+h          
x_last = N_x          


@jit(nopython=True)
def V_sawtooth(x, alpha,N_x):
    '''
    Potential V_1(x) from the task description
    ''' 
    if x <= -(1-alpha)*N_x: 
        x += N_x
    elif  alpha*N_x < x:
        x-= N_x
    if (0 < x <= alpha*N_x):
        return x/(alpha*N_x) 
    else:
        return -1*(x/((1-alpha)*N_x))
   

@jit(nopython=True)  
def V_constant(x, alpha,N_x):
    '''
    Potential V_2(x) from the task description
    '''
    c = 1
    return c
  
@jit(nopython=True)
def J(n_plus,n_minus, N_p):
    '''
    Returns the normalized particle current
    '''
    return (n_plus-n_minus)/N_p


def J_avg(current_list):
    '''
    Returns the average of a list
    '''
    average = sum(current_list)/len(current_list)
    return average

@jit(nopython=True)
def p_minus_cycle(x_0, V, alpha, beta_k, N_x): 
    '''
    Probability of random walk step to the left. 
    Uses periodic boundary conditions
    '''
    value = 0
    x_first = -N_x+h
    x_last = N_x
    if x_0 == x_first:
        value = 1/(1+e**(-beta_k*(V(x_0+1,alpha, N_x)-V(x_last, alpha, N_x)))+e**(-beta_k*(V(x_0,alpha,N_x)-V(x_last,alpha,N_x))))
    elif x_0 == x_last:
        value = 1/(1+e**(-beta_k*(V(x_first,alpha,N_x)-V(x_0-1,alpha,N_x)))+e**(-beta_k*(V(x_0,alpha,N_x)-V(x_0-1,alpha,N_x))))
    else:
        value = 1/(1+e**(-beta_k*(V(x_0+1,alpha,N_x)-V(x_0-1,alpha,N_x)))+e**(-beta_k*(V(x_0,alpha,N_x)-V(x_0-1,alpha,N_x))))
    return value

@jit(nopython=True)
def p_plus_cycle(x_0, V, alpha, beta_k, N_x):
    '''
    Probability of random walk step to the right. 
    Uses periodic boundary conditions
    '''
    x_first = -N_x+h
    x_last = N_x
    value = 0
    if x_0 == x_first:
        value = 1/(1+e**(-beta_k*(V(x_last,alpha, N_x)-V(x_0+1,alpha, N_x)))+e**(-beta_k*(V(x_0,alpha, N_x)-V(x_0+1,alpha, N_x))))
    elif x_0 == x_last:
        value = 1/(1+e**(-beta_k*(V(x_0-1,alpha, N_x)-V(x_first,alpha, N_x)))+e**(-beta_k*(V(x_0,alpha, N_x)-V(x_first,alpha, N_x))))
    else:
        value = 1/(1+e**(-beta_k*(V(x_0-1,alpha, N_x)-V(x_0+1,alpha, N_x)))+e**(-beta_k*(V(x_0,alpha, N_x)-V(x_0+1,alpha, N_x))))
    return value

@jit(nopython=True)
def fill_particles():
    '''
    Returns an array of particles evenly distributed along the x-axis
    '''
    particles = np.zeros(N_p,dtype = np.int64)
    number = x_first
    for i in range(1,len(particles)+1):
        particles[i-1] = number
        if i%6 == 0:
            number += 1
    return particles

@jit(nopython=True)
def get_current_per_time_step(V, init, time_steps, alpha, beta_k, N_p, N_x):
    '''
    Returns an array of the normalized particle current per time step
    '''
    x_first = -N_x+h
    x_last = N_x
    current_list_per_time_step = []
    for step in range(time_steps):
        countPos = 0
        countNeg = 0
        for position in range(len(init)):
            p = np.random.uniform(0,1)
            if p <= p_minus_cycle(init[position],V, alpha, beta_k, N_x):
                if init[position] == x_first:
                    init[position] = x_last
                else:
                    init[position] -= 1
                countNeg +=1
            elif p > (1-p_plus_cycle(init[position],V, alpha, beta_k, N_x)):
                if init[position] == x_last:
                    init[position] = x_first
                else:
                    init[position] += 1
                countPos +=1
        current_list_per_time_step.append(J(countPos, countNeg, N_p))
    return current_list_per_time_step

def get_cycle_averaged_current(particles, V1, V2, T_p, jumps, alpha, beta_k, N_p, N_x):
    '''
    Returns an array of the cycle-averaged currents for the specified jumps (jumps = 2*cycles)
    '''
    average_currents = []
    for i in range(jumps):
        if i%2 == 0:
            current_list = []
            currents_per_time_step = get_current_per_time_step(V2,particles,T_p, alpha, beta_k, N_p, N_x)
            current_list = currents_per_time_step
        else:
            currents_per_time_step = get_current_per_time_step(V1,particles,T_p, alpha, beta_k, N_p, N_x)
            current_list += currents_per_time_step
            average_currents.append(round(J_avg(current_list),15)) #Round the currents to as many digits as possible while the array below is visible in it's entirety on the screen. 
    return average_currents

def plot_potential():
  '''
  Plots the sawtooth potential
  '''
  x = np.linspace(-N_x,N_x,2*N_x)
  y = np.zeros(len(x))
  for i in range(len(x)):
    y[i] = V_sawtooth(x[i],alpha,N_x)
  plt.plot(x,y)
  plt.xlabel("$x$", fontsize = 20)
  plt.ylabel("$V(x)$",fontsize = 20)
  plt.title(r"$\alpha = $" + str(alpha),fontsize = 20)
  plt.show()

@jit(nopython=True)
def reset_positioning(N_p,N_x):
    '''
    Returns an array of particles which are reset to: half of the particles 
    at each of the two x-coordinates which coincide with minima of V_sawtooth. 
    '''
    particles = np.zeros(N_p,dtype = np.int64)
    for i in range(len(particles)):
        if i < (len(particles)/2):
            particles[i] = 0
        else: 
            particles[i] = N_x 
    return particles

def avg_current_vs_time(T_p_values, beta_k, N_p, N_x):    
    '''
    Returns an array of average currents, which can be used to plot vs T_p
    '''
    average_currents = []
    particles = reset_positioning(N_p, N_x)
    for T_p in T_p_values:
        average_currents+=get_cycle_averaged_current(particles, V_sawtooth, V_constant, T_p, jumps, alpha, beta_k, N_p, N_x)
    return average_currents

def avg_current_vs_alpha(alpha_values, beta_k, N_p, N_x):       
    '''
    Returns an array of average currents, which can be used to plot vs alpha
    '''
    average_currents = []
    
    for alpha in alpha_values:
        particles = reset_positioning(N_p, N_x)
        average_currents+=get_cycle_averaged_current(particles, V_sawtooth, V_constant, T_p, jumps, alpha, beta_k, N_p, N_x)
    return np.array(average_currents)

  
```

* Why we get non-zero currents: 

Since we do not have the symmetric case (where $\alpha = 0.5$) there has to be a non-zero current. As we have seen earlier, if the potential is strong enough the particles will always move towards lower potential. In this simulation $\beta k=1000$, which, based on the results in exercise 4, indeed can be said to be strong enough. In that exercise we clearly saw that the particles moved towards the lowest possible potential when $\beta k=100$, which implies that it is fair to expect that the potential dictates the movement of the particles even more at $\beta k=1000$. As we can see in the illustrations of the potentials, when $\alpha =0.8$ the sawtooth potential is leaning to the right, which means that the bottom is offset to the left. In other words, a particle de-exciting into the sawtooth potential will, on average, travel to the left, since it will move towards the bottom to minimize its potential.
When $\alpha = 0.2$ we will have the opposite situation, where the potential  leans leftward, resulting in a positive current. We can also see that the absolute values of the current are approximately equal whether $\alpha=0.8$ or $\alpha = 0.2$. 


* Compare $J_{avg}(0)$ with $J_{avg}(n)$ for n > 0:

It is apparent that the averaged current during the first cycle is bigger in absolute value than the other cycle-averaged currents. The fact that the particles seek to minimize their potential energy can be used to give a physical interpretaion of this observation. The particles are the furthest from the positions of lowest potential in the beginning, when they are uniformly distributed. In contrast, all the consecutive cycles have an initial position where all the particles are in either one of the minimas. With this initial distribution the constant potential only distributes the particles isotropic with center at the minimas, but not quite uniform. This is why in the first cycle the potential have more particles to move in to the minima than in the cycles after.

The two x-coordinates which coincide with the minima of $V_1(x)$ must be $x = 0$ and $x = 0 + nN_x$, since it is periodic. The part of $V_1(x)$ which lies within $0 < x \leq \alpha N_x$ is a monotonically increasing function, which means that it's minima is when $x$ is the smallest in the given interval $\Rightarrow$ $x = 0$. The other part of $V_1(x)$ is a monotonically decreasing function, which means that it's minima is at the highest possible $x$ value in the interval $\Rightarrow$ x = 0. However, because of the periodicity of the function, the other minima is at $x = N_x$.

## Programming exercise 6.1

```python
T_p = 500
jumps = 2
beta_k = 1000
h = 1   #step length
N_x=90  
N_p = 12*N_x
x_first=-N_x+h          
x_last = N_x


from scipy import special

def J_avg_an(alpha,T_p):
  return  N_x/(4*T_p) * (special.erfc( alpha*N_x/2 * np.sqrt(3/T_p) ) - special.erfc( (1-alpha)*N_x/2 * np.sqrt(3/T_p)))

alpha_values = np.linspace(0.01,1,50)   #Alpha can't be zero
plt.plot(alpha_values, avg_current_vs_alpha(alpha_values, beta_k, N_p, N_x), label="Numerical")
plt.plot(alpha_values,J_avg_an(alpha_values,T_p), label="Analytical")
plt.xlabel(r"$\alpha$", fontsize=15)
plt.ylabel(r"$J_{\mathrm{avg}}$", fontsize=15)
plt.legend()
plt.show()
```

### Comparison and comment on the graphs: 

It is apparent that the analtical solution and the numerical solution are strongly coherent. This is in accordance with the assumption, in the analytical solution, that the particles relax towards the potential minima, since we are resetting the positions of the particles to exactly these minimas when running the numerical solution.
